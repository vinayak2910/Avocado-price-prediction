```python
x=df_new
```

```python
y=df_t.iloc[:,0]
```

```python
y.shape
```

```python
#removing skewness
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
```

```python
#scaling using standard scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
#finding the random state
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score

```

```python
best_rstate=0
accu=0

for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    model=LinearRegression()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    tempacc=r2_score(y_test,y_pred)
    if tempacc>accu:
        accu=tempacc
        best_rstate=i
        
print(f"best accuracy is : {accu*100}found on random state: {best_rstate}")      
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=69)
```

```python
def models(x_tarin,y_train,x_test,y_test):
    #using KneighborClassfier mrethod of neighbor class to use nearest neighbor algorith 
    from sklearn.neighbors import KNeighborsClassifier
    knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    knn.fit(x_train,y_train)
    
    
    #using svc 
    from sklearn.svm import SVC
    svc=SVC(kernel='linear')
    svc.fit(x_train,y_train)
    
    
    #Using SVC method of svm class to use Kernel SVM Algorithm
    from sklearn.svm import SVC
    svc_rbf = SVC(kernel = 'rbf')
    svc_rbf.fit(x_train, y_train)
    #svc_rbf.predict(x_test,y_test)

  #Using GaussianNB method of naïve_bayes class to use Naïve Bayes Algorithm
    from sklearn.naive_bayes import GaussianNB
    gauss = GaussianNB()
    gauss.fit(x_train, y_train)
    #gauss.predict(x_test,y_test)

  #Using DecisionTreeClassifier of tree class to use Decision Tree Algorithm
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion = 'entropy')
    tree.fit(x_train, y_train)
    #tree.predict(x_test,y_test)

  #Using RandomForestClassifier method of ensemble class to use Random Forest Classification algorithm
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10, criterion = 'entropy')
    forest.fit(x_train, y_train)
    #forest.predict(x_test,y_test)
  
  #print model accuracy on the training data
    print('[0]K Nearest Neighbor Training Accuracy:', knn.score(x_train, y_train))
    print('[0]K Nearest Neighbor Testing Accuracy:', accuracy_score(y_test,knn.predict(x_test)))
    print()
    print('[1]Support Vector Machine (Linear Classifier) Training Accuracy:', svc.score(x_train, y_train))
    print('[1]Support Vector Machine (Linear Classifier) Testing Accuracy:', accuracy_score(y_test,svc.predict(x_test)))
    print()
    print('[2]Support Vector Machine (RBF Classifier) Training Accuracy:', svc_rbf.score(x_train, y_train))
    print('[2]Support Vector Machine (RBF Classifier) Testing Accuracy:', accuracy_score(y_test,svc_rbf.predict(x_test)))
    print()
    print('[3]Gaussian Naive Bayes Training Accuracy:', gauss.score(x_train, y_train))
    print('[3]Gaussian Naive Bayes Testing Accuracy:', accuracy_score(y_test,gauss.predict(x_test)))
    print()
    print('[4]Decision Tree Classifier Training Accuracy:', tree.score(x_train, y_train))
    print('[4]Decision Tree Classifier  Testing Accuracy:', accuracy_score(y_test,tree.predict(x_test)))
    print()
    print('[5]Random Forest Classifier Training Accuracy:', forest.score(x_train, y_train))
    print('[5]Random Forest Classifier  Testing Accuracy:', accuracy_score(y_test,forest.predict(x_test)))
  
    return knn, svc, svc_rbf, gauss, tree, forest
```
